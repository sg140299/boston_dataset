```python
import numpy as np
import pandas as pd
def run():
    data=np.loadtxt("train_.csv",delimiter=',')
    df=pd.DataFrame(data)
    
    
    
    iteration=1000
    learning_rate=0.13
    
    
    m=gd(data,learning_rate,iteration)
    return m
def cost(x,y,m):
    return (1/len(x))*(((y-((x*m).sum(axis=1)))**2).sum())
def gd(points,learning_rate,iteration):
    m=np.zeros(points.shape[1])
    y_train=points[:,-1]
    
    
    x_train=points[:,:-1]

    #x_train=arg.fit_transform(x_train)
    temp=np.ones((points.shape[0],points.shape[1]))
    temp[:,:-1]=x_train           
    x_train=temp
   
    for i in range(iteration):
        m=step_gradient(x_train,y_train,learning_rate,m)
     
        print(cost(x_train,y_train,m))
    return m
def step_gradient(x_train,y_train,learning_rate,m):
    slope_m=np.zeros(len(m))
   
    
    n=len(x_train)

    for i in range(n):
        
        for j in range(len(m)):
            slope_m[j]+=(-2/n)*((y_train[i]-(x_train[i]*m).sum())*x_train[i][j])
        
       
    new_m=m-learning_rate*slope_m
    
    return new_m
m=run()

#print(m)
test_data=np.genfromtxt('test_.csv',delimiter=',')
#test_data=arg.transform(test_data)
amp=np.ones((test_data.shape[0],test_data.shape[1]+1))
amp[:,:-1]=test_data
test_data=amp

def pred(m,test_data):
    return (m*test_data).sum(axis=1)
        
    
a=pred(m,test_data)
np.savetxt('prediction2.csv',a,delimiter=',')
#print(a)
```
